# 02. Merging the products and previously merged orders datasets

```python
#Dropping merge check column in merged orders dataset

df_ords_merged = df_ords_merged.drop(columns=[col for col in df_ords_merged.columns if '_merge' in col])
```

```python
#Merging the producst and previously merged orders datasets

df_ords_prods_merged = df_prods.merge(df_ords_merged, on = 'product_id', indicator = True)
```
