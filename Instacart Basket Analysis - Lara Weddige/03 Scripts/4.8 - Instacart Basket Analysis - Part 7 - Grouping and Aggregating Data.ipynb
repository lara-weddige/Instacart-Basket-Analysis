# 03. Finding differences in spending habits

- Creating max order column
- Creating loyalty flga
- Looking at differences in spending habits

```python
# Splitting data into groups based on user_id column
# Generating maximum order info for each customer based on order_number column
# Creating a max order column 

ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Checking for new max_order column 

ords_prods_merge.head()
```

```python
#Assigning loyalty flags to customers based on order numbers

ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Printing frequency of new loyalty_flag column

ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
# Determining whether prices of products purchased by loyal customers differ from those regular or new customers

ords_prods_merge.groupby('loyalty_flag').agg({'prices': ['mean', 'min', 'max']})
```

# 04. Different types of spenders

- Creating average price per order column
- Creating a spending flag

```python
# Creating an average_price_order column

ords_prods_merge['av_price_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge.head()
```

```python
#Creating a spending flag for each user based on the average price across all their orders

ords_prods_merge.loc[ords_prods_merge['av_price_order'] >= 10, 'spending_flag'] = 'High spender'
ords_prods_merge.loc[ords_prods_merge['av_price_order'] < 10, 'spending_flag'] = 'Low spender'
```

```python
#Counting

ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

# 05. Determining frequency of orders

- Creating order regularity column
- Assigning order frequency flag

```python
# Creating an order regularity column

ords_prods_merge['order_regularity'] = ords_prods_merge.groupby(['user_id'])['days_since_last_order'].transform(np.median)
```

```python
ords_prods_merge.head()
```

```python
#Assigning order frequency flags to customers based on order numbers

ords_prods_merge.loc[ords_prods_merge['order_regularity'] > 20, 'order_frequency_flag'] = 'Non-frequent customer'
ords_prods_merge.loc[(ords_prods_merge['order_regularity'] <= 20) & (ords_prods_merge['order_regularity'] > 10), 'order_frequency_flag'] = 'Regular customer'
ords_prods_merge.loc[ords_prods_merge['order_regularity'] <= 10, 'order_frequency_flag'] = 'Frequent customer'
```

```python
#Counting

ords_prods_merge['order_frequency_flag'].value_counts(dropna = False)
```
